```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60,40,30,70]
labels = ['Sixty','Forty','Extra1','Extra2']
colors = ['Red','Blue','Green','Orange']
plt.pie(slices,labels = labels,colors = colors,wedgeprops = {'edgecolor': 'Yellow'})
plt.title('My Fisrt Pie Chart')
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


# Language Popularity

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ['Red','Blue','Green','Orange','Magenta']
explode = [0,0,0,0.1,0]

plt.pie(slices,labels = labels,explode = explode,startangle= 90,shadow = True,autopct = '%1.1f%%',colors = colors,wedgeprops = {'edgecolor': 'Yellow'})

plt.title('My Fisrt Pie Chart')
plt.show()
```
