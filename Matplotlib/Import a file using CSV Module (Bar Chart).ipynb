```python
from matplotlib import pyplot as plt
import numpy as np
import csv
from collections import Counter
                                                     # can be done with pandas to read csv # import pandas as pd
with open('data.csv') as csv_file:               # data = pd.read_csv('data.csv')
    csv_reader = csv.DictReader(csv_file)        # ids = data['responder_id'] # lang_responses = data['LanguagesWorkedWith']
    language_counter = Counter()                
#     row = next(csv_reader)   # {'Responder_id': '1', 'LanguagesWorkedWith': 'HTML/CSS;Java;JavaScript;Python'}
#     print(row['LanguagesWorkedWith'].split(';')) # ['HTML/CSS', 'Java', 'JavaScript', 'Python']
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
# print(language_counter,'\n')
# print(language_counter.most_common(15))
    
languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# print(languages)
# print(popularity)
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)


plt.title('Most Popular Languages')
# plt.ylabel('Programming Languages')
plt.xlabel('No.of Peoples used')

plt.show()
```
